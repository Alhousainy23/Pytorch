## Introduction to tensors

Now we've got PyTorch imported, it's time to learn about tensors.

Tensors are the fundamental building block of machine learning.

Their job is to represent data in a numerical way.

For example, you could represent an image as a tensor with shape `[3, 224, 224]` which would mean `[colour_channels, height, width]`, as in the image has `3` colour channels (red, green, blue), a height of `224` pixels and a width of `224` pixels.

![example of going from an input image to a tensor representation of the image, image gets broken down into 3 colour channels as well as numbers to represent the height and width](https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/00-tensor-shape-example-of-image.png)

In tensor-speak (the language used to describe tensors), the tensor would have three dimensions, one for `colour_channels`, `height` and `width`.

But we're getting ahead of ourselves.

Let's learn more about tensors by coding them.

```python
# Create a tensor 
r_tensor = torch.tensor(data=[1,2,3,4,5,6,7], dtype=torch.float32,
                        device=None, requires_grad=True)
print("The Random Tensor is: ",r_tensor,
        "\nThe Rank of the Random Tensor is: ",r_tensor.ndim ,
        "\nThe Shape of the Random Tensor is " , r_tensor.shape,
        "\nThe Data Type of the Random Tensor is: ",r_tensor.dtype, 
        "\nThe Device of the Random Tensor is: ",r_tensor.device,
        "\nThe Requires Grad of the Random Tensor is: ",r_tensor.requires_grad)
```

```python
# Tensors Operations[Multiply , addtition - subtraction - division ]
x_tensor1 = torch.tensor(data=[1,2,3,4,5,6,7])
# First operation is addtition 
x_tensor1 += 10 
print("The Tensor after Addition is: ",x_tensor1)
#===============================================================================
#Second operation is subtraction
x_tensor1 -= 2
print("The Tensor  after Subtraction is : ", x_tensor1)
#===================================================================================
# Third operation Is Multiplication 
x_tensor1 *=2
print("The Tesnor after Multiply is : ", x_tensor1)
#=================================================================================
# Fourth Operation is Division 
x_tensor1 = x_tensor1 / 2
print("The Tensor after Divison is : ", x_tensor1)
```

```python
# Tensors Operations[Multiply , addtition - subtraction - division ]
x_tensor1 = torch.tensor(data=[1,2,3,4,5,6,7])
# First operation is addtition 
print("The Tensor after Addition is: ",torch.add(x_tensor1,10))
#===============================================================================
#Second operation is subtraction
print("The Tensor  after Subtraction is : ", torch.sub(x_tensor1,2))
#===================================================================================
# Third operation Is Multiplication 
print("The Tesnor after Multiply is : ",torch.multiply(x_tensor1,10))
#=================================================================================
# Fourth Operation is Division 
print("The Tensor after Divison is : ", torch.div(x_tensor1,2))
```

```python
# Matrix Multiplication by using torch.matmul [ matrix multiplication]
matrix = torch.matmul(x_tensor1,x_tensor1)
print("The Tensor after applu matrix multiplication is : ", matrix,
      "\n The Shape Matrix after Matmul is : ", matrix.shape,
      "\n The Rank of matrix after MatMul is : ", matrix.ndim,
      "\n The date type matrix after MatMul is : ", matrix.dtype)
```

```python
#device agnostic code 
device = 'cuda' if torch.cuda.is_available() else 'cpu' 
print("This device is ==> ", device)
print("The number of GPUs for in this device is ==> ", torch.cuda.device_count())
#Check tensor on which device 
tensor4 = torch.randn([2,5])
print("The data at the tensor4 are ==> ", tensor4,
      "\nThe tensor4 on this device is by default ==> ", tensor4.device)
#I need to put this tensor on Cuda [GPU] 
tesnor_gpu = tensor4.to('cuda')
print("The tensor GPU device is ==> ", tesnor_gpu.device)
#Another way by using agnostic code 
tensor4 = tensor4.to(device)
print("The tensor4 after send to GPU ==> ", tensor4 ,
      "\nThe tensor4 device is ==> ", tensor4.device)
```

```python
#IF i need to return Tensor4 run on CPU Device instead of GPU
cpu_tensor = tensor4.cpu()
print("The tensor4 after recover to cpu ==> ", cpu_tensor ,
      "\nThe tensor4 type is ==> ",type(cpu_tensor),
      "\nThe tensor4 device is ==> ", cpu_tensor.device)
print("======="*10)
#And i can change to numpy after run on cpu 
cpu_tensor = tensor4.cpu().numpy()
print("The tensor4 after recover to cpu ==> ", cpu_tensor,
      "\nThe tensor4 after convert to Numpy type is ==> ", type(cpu_tensor),
      "\nThe tensor4 device is ==> ", cpu_tensor.device)
```
